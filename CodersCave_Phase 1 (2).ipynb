# Data cleaning

```python
cols = df.columns
cols
```

```python
df.drop_duplicates() 
```

```python
int_cols=df.select_dtypes(include=np.number).columns.tolist()
for i in int_cols:
    df[i] = df[i].fillna(df[i].median())
```

```python
df.isnull().sum()
```

```python
city_wise=df['City'].value_counts().reset_index()
city_wise.rename(columns={"index":'City Name','City':'count'}, inplace=True)
city_wise
```

# Top 10 Attacked Cities

```python
import seaborn as sns
ax = sns.barplot(x='count', y="City Name", data=city_wise[:10],
                 palette="Greens_d").set_title('Top 10 Attacked cities')
```

The graph shows that Unknown and Baghdad are the top most attacked cities.

# Attack Type Frequency Analysis 

```python
attack_type=df['AttackType'].value_counts().reset_index()
attack_type.rename(columns={'index':'AttackType','AttackType':'count'},inplace=True)
attack_type
```

# Top 10 Attack Type

```python
ax = sns.barplot(x="count", y='AttackType', data=attack_type,
                 palette="Blues_d").set_title('Most used attack type')
```

Bombing/explosion is the most common type of terrorist attack.It also shows that armed assault and assassination are two of the next most common types of attacks.

# Yearly Analysis of Total Wounded Individuals using Line Plot

```python
Wounded_by_year = df[["Year","Wounded"]].groupby("Year").aggregate(np.sum)
Wounded_by_year.plot(color = 'b');
```

The number of wounded was lower before 2000 and increased after 2000.

```python
#no. of terror attacks
plt.figure(figsize=(10,6))
fig = sns.distplot(df["Year"].values, color = 'r')
fig.set_xlabel(".Year",size=10)
fig.set_ylabel("Terror attack total",size=10)
plt.title('Number of Terror attacks each year',size = 15)
plt.show()
```

```python
import plotly.express as px

#to avoid lagging, only show the first 1k instead of the 180,000 data points.
geog = df.head(1000)
geog = geog[['Latitude','Longitude']]
fig = px.scatter_geo(geog, lat='Latitude', lon='Longitude')

fig.show()
```

Yes, we can visually identify the regions that are most targeted by leveraging Plotly for visualization.By mapping the data, we can observe which parts of the world are most frequently targeted in terms of terror attacks.

```python
from wordcloud import WordCloud

# Generate a word cloud of targeted places
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(df['Target_type'].astype(str)))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Targeted Places')
plt.show()
```

# Visualization of Wounded casualty by Region

```python
wounded_by_region = df.groupby('Region')['Wounded'].sum().sort_values(ascending=False)
plt.figure(figsize=(14, 8))
wounded_by_region.plot(kind='bar', stacked=True)
plt.title('Stacked Bar Chart of wounded by Region')
plt.xlabel('Region')
plt.ylabel('Total count of wounded casualty')
plt.show()
```
